```python
from openai import OpenAI
import os
from google.colab import userdata
my_key = userdata.get('OPENAI_API_KEY')
os.environ['OPENAI_API_KEY'] = my_key
client = OpenAI()

response = client.embeddings.create(
    input="hi how are you",
    model="text-embedding-3-small"
)

print(response.data[0].embedding)
```

```python
input_tokens = tokenizer("explain the deep learning in depth", return_tensors = 'pt')
input_tokens
```

```python
import torch
from transformers import AutoModelForCausalLM

model = AutoModelForCausalLM.from_pretrained("gpt2")
```

```python
out = model(input_ids=input_tokens["input_ids"])
out
```

```python
gen_out = model.generate(input_ids = input_tokens["input_ids"], max_new_tokens=20)
gen_out
```

```python
tokenizer.batch_decode(gen_out)
```
